```python
def linear(xdata, m, b):
    return m*xdata + b

def find_nearest_index(arr, target):
    """
    Find the index of the element in the array 'arr' that is nearest to the 'target' value.
    
    Args:
        arr (list): The list of values.
        target (float): The value to which you want to find the nearest value.
    
    Returns:
        int: The index of the nearest element in the array.
    """
    # Calculate the absolute differences between the elements and the target value.
    absolute_diff = [abs(x - target) for x in arr]
    
    # Find the index of the minimum absolute difference.
    nearest_index = absolute_diff.index(min(absolute_diff))
    
    return nearest_index
```

```python
data = pd.read_csv(r'/home/lewis/Documents/Honours_Research/data/csv/metalicity_fit/gamma_z.csv')
metalicity = [0.03,0.0142,0.00142,0.000142]
std_gamma = [0.01214251775481037,0.012716992782249017,0.013631714303811406,0.020455631234779774]
log_m = np.log10(metalicity)
domain = np.linspace(log_m[0],log_m[-1],50)
popt,pcov = curve_fit(linear,log_m,data.gamma_z)
plt.figure(figsize = (3.5, 2.5))
plt.scatter(log_m,data.gamma_z,label = 'model')
plt.errorbar(log_m,data.gamma_z, yerr = std_gamma, fmt = 'none', capsize = 3)
plt.plot(domain,linear(domain,*popt), label = 'fit')
plt.xlabel('$\log{Z}$')
plt.ylabel('$M_\mathrm{inter,f}/M_\mathrm{core,f}$')
plt.legend(fontsize = 'small')
plt.savefig('/home/lewis/Documents/Honours_Research/data/plots/Plots_for_paper/metalicity/gamma_fit.pdf', bbox_inches = 'tight')

print(linear(log_m,*popt))
print(popt)
```
